## Esercizio 3 - Rete neurale custom: predizione della temperatura critica dei superconduttori

### Dataset: Superconductivity Data (UCI)

Un materiale e' definito *superconduttore* quando, al di sotto di una certa temperatura detta **temperatura critica (Tc)**, perde completamente la resistenza elettrica. Prevedere la Tc di un materiale a partire dalla sua composizione chimica e' un problema aperto e rilevante nella fisica della materia condensata e nella scienza dei materiali.

Il dataset contiene informazioni su **21.263 superconduttori**. Per ciascun materiale sono disponibili **81 feature** derivate dalle proprieta' atomiche degli elementi costituenti (media, media pesata, deviazione standard, entropia, gamma, ecc. di: massa atomica, raggio atomico, densita', affinita' elettronica, calore di fusione, conducibilita' termica, valenza). Il target e' la temperatura critica in Kelvin.

Fonte: https://archive.ics.uci.edu/dataset/464/superconductivty+data

File di dati: `data/train.csv` (principale, 21263 righe x 82 colonne, ultima colonna = `critical_temp`).
