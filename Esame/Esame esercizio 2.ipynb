## Esercizio 2 - Analisi di serie storica: Sunspot Numbers

### Dataset: Monthly Sunspot Number (SILSO)

Il numero di macchie solari (sunspot number) e' uno degli indicatori piu' antichi e meglio studiati dell'attivita' solare: misura il numero di regioni scure visibili sulla superficie del Sole, legate a intensa attivita' magnetica. La serie presenta un ciclo ben noto di circa **11 anni** (ciclo di Schwabe), con variazioni in ampiezza di lungo periodo (cicli di Gleissberg ~90 anni) e una stagionalita' apparente dovuta al moto di rivoluzione terrestre.

I dati ufficiali sono pubblicati dal **Royal Observatory of Belgium - SIDC/SILSO (Sunspot Index and Long-term Solar Observations)** e coprono il periodo dal 1749 ad oggi, con oltre 3.300 osservazioni mensili.

Fonte: https://www.sidc.be/SILSO/datafiles

File di dati: `data/SN_m_tot_V2.0.csv` (colonne: anno, mese, frazione decimale, SNvalue, SNerror, Nb observations, definitive/provisional flag).

La variabile di interesse e' **SNvalue**: numero medio mensile di macchie solari.
