## Esercizio 1 - Classificazione e Anomaly Detection

### Dataset: MAGIC Gamma Telescope

Il dataset proviene dal telescopio **MAGIC (Major Atmospheric Gamma Imaging Cherenkov)**, installato alle Isole Canarie. Si tratta di un telescopio terrestre che sfrutta la tecnica Cherenkov atmosferica: quando un fotone gamma ad alta energia entra nell'atmosfera, genera uno sciame di particelle secondarie che emettono luce Cherenkov, registrata come un'immagine ellittica nel piano focale del telescopio. Lo stesso tipo di immagine viene però prodotto anche da raggi cosmici adronici (protoni, nuclei), che costituiscono il rumore di fondo.

Il dataset (in forma simulata via Monte Carlo) contiene **19.020 eventi**, ciascuno descritto da 10 feature continue che caratterizzano la forma dell'ellisse dell'immagine Cherenkov:

- `fLength`: lunghezza dell'asse maggiore dell'ellisse [mm]
- `fWidth`: lunghezza dell'asse minore dell'ellisse [mm]
- `fSize`: log10 della somma del contenuto dei pixel [#phot]
- `fConc`: rapporto tra la somma dei due pixel piu' brillanti e fSize
- `fConc1`: rapporto tra il pixel piu' brillante e fSize
- `fAsym`: distanza del pixel piu' brillante dal centro, proiettata sull'asse maggiore [mm]
- `fM3Long`: terza radice del terzo momento lungo l'asse maggiore [mm]
- `fM3Trans`: terza radice del terzo momento lungo l'asse minore [mm]
- `fAlpha`: angolo dell'asse maggiore con il vettore dall'origine [gradi]
- `fDist`: distanza dall'origine al centro dell'ellisse [mm]

**Target (Attributo 11):** classe dell'evento
- `g` = gamma (signal)
- `h` = hadron (background)

Il dataset e' disponibile su UCI Machine Learning Repository:
https://archive.ics.uci.edu/dataset/159/magic+gamma+telescope
